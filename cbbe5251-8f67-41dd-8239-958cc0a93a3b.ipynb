```python
def remove_outliers_percentile(data, lower_percentile=0.05, upper_percentile=0.95):
    filtered_data = data.copy()
    filtered_data = filtered_data.ffill()
    
    mask = np.ones(len(filtered_data), dtype=bool)
    
    for column in filtered_data.select_dtypes(include=[np.number]).columns:
        nan_percentage = filtered_data[column].isna().mean()
        if nan_percentage > 0.5:
            continue

        valid_data = filtered_data[column].dropna()
        if len(valid_data) == 0:            
            continue
            
        lower_bound = np.quantile(valid_data, lower_percentile)
        upper_bound = np.quantile(valid_data, upper_percentile)

        column_mask = (filtered_data[column] >= lower_bound) & (filtered_data[column] <= upper_bound)
        mask = mask & (column_mask | filtered_data[column].isna())

    return filtered_data[mask]
gold_train = remove_outliers_percentile(gold_train)
```

```python
def smape(y_true, y_pred):
    n = len(y_true)
    smape_values = []
    for actual, predicted in zip(y_true, y_pred):
        numerator = abs(actual - predicted)
        denominator = (abs(actual) + abs(predicted)) / 2
        smape_values.append(numerator / denominator if denominator != 0 else 0.0)
    return (100 / n) * sum(smape_values)

def final_smape(rougher_smape, final_smape):
    final = 0.25 * rougher_smape + 0.75 * final_smape 
    return final
```

```python
def split_scale_train(data, model, feature, param_distributions, n_iter=10, cv=5):
    data_train, data_val = train_test_split(data, test_size=0.25, random_state=12345)
    
    features_train = data_train.drop(feature, axis=1)
    target_train = data_train[feature]
    features_valid = data_val.drop(feature, axis=1)
    target_valid = data_val[feature]


    random_search = RandomizedSearchCV(
        estimator=model,
        param_distributions=param_distributions,
        n_iter=5, 
        cv=3,   
        scoring='neg_root_mean_squared_error',
        random_state=12345,
        n_jobs=2 
    )
    
    random_search.fit(features_train, target_train)
    
    predictions = random_search.best_estimator_.predict(features_valid) 
    rmse = np.sqrt(mean_squared_error(target_valid, predictions))
    
    print("Best parameters:", random_search.best_params_) 
    print("Best cross-validation score:", abs(random_search.best_score_))  
    print("Validation RMSE:", rmse)
        
    return predictions, rmse, random_search.best_estimator_
```
